```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Student_Performance.csv"

# Load the latest version
df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "nikhil7280/student-performance-multiple-linear-regression",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

df.head()
```

```python
def col_formatter(cols):
    col_list = []
    for x in list(cols):
        x = x.replace(" ","_")
        col_list.append(x)
    return col_list

df.columns = col_formatter(df.columns)
df.head()
```

```python
import pandas as pd

dummies = pd.get_dummies(df[['Extracurricular_Activities']], dtype='int')
dummies
```

```python
df.drop(str(df.columns[2]), axis=1, inplace=True)
df.head()
```

```python
df = pd.concat([df, dummies], axis=1)
df.head()
```

```python
x,y = df.drop(str(df.columns[4]),axis=1), df[['Performance_Index']]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_scaled = scaler.fit_transform(x_train)
test_scaled = scaler.fit_transform(x_test) 
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
result = model.fit(x_train, y_train)
```

```python
test_results = result.predict(x_test)
test_results
```

```python
from sklearn import metrics

print(f"MAE: {metrics.mean_absolute_error(y_test, test_results)}")
print(f"MSE: {metrics.mean_squared_error(y_test, test_results)}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_test, test_results))}")
print(f"r2_score: {metrics.r2_score(y_test, test_results)}")
```
